## 1. Supervised Learning – Regression

**Goal:** Predict a numerical value.

**Example:** Predict house price from area.

We already know the *correct prices* during training.

```python
import numpy as np
from sklearn.linear_model import LinearRegression

areas = np.array([50, 70, 90, 110, 130])
prices = np.array([25000, 35000, 45000, 55000, 65000])

X = areas.reshape(-1, 1)
y = prices

model = LinearRegression()
model.fit(X, y)

print("Predicted price for 100 m²:", model.predict([[100]])[0])
```

## 2. Supervised Learning – Classification

**Goal:** Predict a category or class.

**Example:** Will it rain or not?

- 1 → Rain
- 0 → No Rain

```python
from sklearn.linear_model import LogisticRegression

# Features: humidity (%), temperature (°C)
X = np.array([
    [80, 20],
    [90, 18],
    [60, 25],
    [30, 30],
    [85, 19]
])

# Labels: 1 = Rain, 0 = No Rain
y = np.array([1, 1, 0, 0, 1])

clf = LogisticRegression()
clf.fit(X, y)

prediction = clf.predict([[75, 22]])
print("Will it rain?", "Yes" if prediction[0] == 1 else "No")
```

## 3. Unsupervised Learning – Clustering

**Goal:** Discover patterns without labels.

**Example:** Group houses by size and price.

No correct answers are given to the algorithm.

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

data = np.array([
    [50, 25000],
    [60, 30000],
    [120, 60000],
    [130, 65000],
    [80, 40000]
])

kmeans = KMeans(n_clusters=2, random_state=0)
clusters = kmeans.fit_predict(data)

plt.scatter(data[:,0], data[:,1], c=clusters)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("House Clusters")
plt.show()
```

## Final Comparison

- **Regression:** predict numbers (house prices)
- **Classification:** predict classes (rain / no rain)
- **Clustering:** find groups (no labels)

These three ideas cover **most ML problems** you will encounter.
